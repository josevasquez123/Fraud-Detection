### ML MODEL

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

#### SPLIT AND OVERSAMPLING DATA

```python
x = df.drop('Class', axis=1)
y= df['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
#UMBALANCED DATA

y_train.value_counts()
```

```python
sm = SMOTE(sampling_strategy='minority')
```

```python
x_train_res, y_train_res = sm.fit_resample(x_train, y_train)
```

```python
#AFTER OVERSAMPLING, DATA IS BALANCED

y_train_res.value_counts()
```

#### LOGISTIC MODEL

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(x_train_res, y_train_res)
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
print(metrics.classification_report(y_test, y_pred))
```

```python
print(metrics.confusion_matrix(y_test,y_pred))
```

#### RANDOM FOREST MODEL

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=20)
```

```python
rf_model.fit(x_train_res, y_train_res)
```

```python
y_pred_rf = rf_model.predict(x_test)
```

```python
print(metrics.classification_report(y_test, y_pred_rf))
```

```python
metrics.confusion_matrix(y_test, y_pred_rf)
```

PRECISION, RECALL AND F1-SCORE FOR 1 CLASS (UNBALANCED DATA) PRESENT A GOOD VALUES\
IN ORDER OF THAT WE CAN ACCEPT THIS MODEL TO MAKE FUTURE PREDICTIONS

### DELET OUTLIERS FROM DATA

```python
from datasist.structdata import detect_outliers
```

```python
df_columns = df.drop(columns=['Class', 'Time', 'Amount'], axis=1)
df_columns.columns
```

```python
outlier_indices = detect_outliers(df,0,df_columns.columns)
df_no_outliers = df.drop(outlier_indices)
```

```python
for index, column_value in  enumerate(df_no_outliers.drop(columns=['Time','Amount','Class'])):
    plt.figure(index)
    plt.title(f'{column_value} Box Plot')
    sns.boxplot(data=df_no_outliers, x='Class', y=column_value)
```

AS YOU CAN SEE, BOXPLOTS SHOW THAT WE REMOVE OUTLIERS

### ML MODEL

#### SPLIT AND OVERSAMPLING DATA

```python
df_no_outliers['Class'].value_counts()
```

```python
x = df_no_outliers.drop('Class', axis=1)
y = df_no_outliers['Class']
```

```python
y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
sm = SMOTE(sampling_strategy='minority')
```

```python
x_train_res, y_train_res = sm.fit_resample(x_train, y_train)
```

THERE ARE FEW CLASS VALUES AS "1", FOR THAT REASON WE GET A FAIL MODELS \
WE HAVE 15 VALUE AS CLASS "1",CONSEQUENTLY WE HAVE A VERY LITTLE Y_TEST VALUES AND THAT IS NOT ENOUGH TO GET A RELIABLE PREDICT VALUES  

#### RANDOM FOREST MODEL

```python
rf_model = RandomForestClassifier(n_estimators=20)
```

```python
rf_model.fit(x_train_res, y_train_res)
```

```python
y_pred_rf = rf_model.predict(x_test)
```

```python
print(metrics.classification_report(y_test, y_pred_rf))
```

```python
metrics.confusion_matrix(y_test, y_pred_rf)
```

### CONCLUSION

- RANDOM FOREST MODEL GETS THE BEST RESULTS
- WE USE RANDOM FOREST MODEL BEFORE DROP OUTLIERS
- DROP OUTLIERS ALSO DROP Y CLASS VALUES
